```python
import csv
import nltk
import string
import re
import numpy as np
import spacy
nlp = spacy.load('de_core_news_sm')
```

```python
def clean(text):
    text = text.lower()
    text_out=[]
    text = nltk.sent_tokenize(text)
    
    for sent in text:
        words = nltk.word_tokenize(sent)
        words = [x for x in words if x not in string.punctuation]
        words = [x for x in words if x not in ["``", "''", "€", "$", ""]]
   
        words = [x for x in words if x.isalpha()]
        words = [x for x in words if len(x) >1]
  
        
        
        doc = nlp(" ".join(words)) 
       
        text_out += [token.lemma_ for token in doc]
    print(text_out)
    
   

    return text_out
```

```python


hatespeech=""

with open("hatespeech/refugees.csv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter=",")
    csv_hate = list(csv_reader_object)[1:]
    for row in csv_hate:
        x = row[0].replace("#", "")
        x = re.sub(r"http\S+|\.\.+|\S+\/\/+|www\S+|\d+|@[A-Za-z0-9]+","", x)
        hatespeech += " " + x
```

```python
hatespeech = clean(hatespeech)
```
