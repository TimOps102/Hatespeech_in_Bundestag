```python
import csv
import nltk
import string
import re
import numpy as np
import pickle
import spacy
nlp = spacy.load('de_core_news_sm')
```

```python
def clean(text):
    tokenized_sents=[]
    text = nltk.sent_tokenize(text)
    for sent in text:
        sent = nltk.word_tokenize(sent)
        sent = [x for x in sent if x not in string.punctuation]
        sent = [x for x in sent if x not in ["``", "''", "€", "$"]]
        doc = nlp(" ".join(sent),disable = ['perser','ner']) 
       
        text_out = [token.lemma_ for token in doc]
        
        if len(text_out) > 1:
            tokenized_sents.append(text_out)
        
    
    return tokenized_sents
```

```python
def remove_emoji(string):   ## from https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python


hatespeech=""

with open("hatespeech/refugees.csv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter=",")
    csv_hate = list(csv_reader_object)[1:]
    for row in csv_hate:
        x = row[0].replace("#", "")
        x = re.sub(r"http\S+|\.\.+|\S+\/\/+|www\S+|\d+|@[A-Za-z0-9]+|[``''€$]","", x)
        x = x.lower()
        hatespeech += " " + x
```

```python
def create_ngrams(corpora):
    ngram_dict = {}
    #stopwords = nltk.corpus.stopwords.words("german")
    stopwords=["findbecci","--","..", "§", '”','|lbr|', '’', "rebecca", "freeflorian", "florian", "„"]
    for i in range(2):
        ngram_dict[f"{i+2}-grams"] =[]
  
        store = []
        for sents in corpora:

            ngram_dict[f"{i+2}-grams"] += [gram for gram in list(nltk.ngrams(sents,i+2)) if len(set(stopwords).intersection(set(gram)))== 0 and (len(gram[0])>1) and (len(gram[1])>1)]
    
    return ngram_dict
```

```python
ngram_dict = create_ngrams(hatespeech_corpora)
```

```python
def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
save_obj(ngram_dict, "ngrams_hatespeech_with_stopwords_lemmatized")
```
