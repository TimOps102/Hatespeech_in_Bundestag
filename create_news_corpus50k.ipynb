```python
import re
import csv
import nltk
import string
```

```python
news_text=""

with open("news/sentences100k.txt") as f:
    head = [next(f) for x in range(50000)]
    for row in head:
        x = row.split("\t")[1]
        news_text += " " + x
    
```

```python
news_text[:1000]
```

```python
def num_there(s):
    return any(i.isdigit() for i in s)
```

```python
def clean(text):
    text = nltk.word_tokenize(text)
    text = [x for x in text if x not in string.punctuation]
    text = [x for x in text if x not in ["``", "''", "€", "$", ""]]
    text = [x.lower() for x in text]
    #text = [x for x in text if x.isalpha()]
    text = [x for x in text if not num_there(x)]
    text = [x for x in text if len(x) >1 ]
    return text
```

```python
news_list = clean(news_text)
```

```python
len(news_list)
```

```python
filtered_news=[]

for word in news_list:
    if word not in nltk.corpus.stopwords.words("german"):
        filtered_news.append(word)
        
```

```python
print(filtered_news[:1000])
```

```python
textfile = open("news/news_corpus50k.txt", "w")
for element in filtered_news:
    textfile.write(element + "\n")
textfile.close()
```
