```python
import re
import csv
import nltk
import string
```

```python
news_text=""

with open("news/sentences100k.txt") as f:
    head = [next(f) for x in range(50000)]
    for row in head:
        x = row.split("\t")[1]
        x = re.sub(r"http\S+|\.\.+|\S+\/\/+|www\S+|\d+.|@[A-Za-z0-9]+|[``''€$]|\n|%","", x)
        news_text += " " + x
    
```

```python
textfile = open("news/news_text100k.txt", "w")
textfile.write(news_text)
textfile.close()
```
