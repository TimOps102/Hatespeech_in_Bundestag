```python
import csv
import nltk
import string
import re
import numpy as np
import pickle
```

```python
substring=r"http\S+|\.\.+|\S+\/\/+|www\S+|\d+|@[A-Za-z0-9]+|[``''€$]|#|\n|OFFENSE|OTHER|ABUSE|INSULT|PROFANITY|@[_A-Za-z0-9]+|\t|[()]|.lbr."
```

```python
def remove_emoji(string):   ## from https://gist.github.com/slowkow/7a7f61f495e3dbb7e3d767f97bd7304b
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python


hatespeech=""

with open("hatespeech/refugees.csv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter=",")
    csv_hate = list(csv_reader_object)[1:]
    for row in csv_hate:
        x = row[0].replace("#", "")
        x = re.sub(substring,"", x)
        x = x.lower()
        hatespeech += " " + x
```

```python

hatespeech2=""

with open("hatespeech/posts.csv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter=",")
    csv_hate = list(csv_reader_object)[1:]
    for row in csv_hate:
        x = row[2]
        x = re.sub(substring,"", x)
        x+="."
        x = x.lower()
       
        hatespeech2 += " " + x
```

```python

hatespeech3=""

with open("hatespeech/comments.csv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter=",")
    csv_hate = list(csv_reader_object)[1:]
    for row in csv_hate:
        x = row[3]
        x = re.sub(substring,"", x)
        x+="."
        x = x.lower()
     
     
        hatespeech3 += " " + x
```

```python
hatespeech4=""

with open("hatespeech/german_dataset.tsv") as csvdatei:
    csv_reader_object = csv.reader(csvdatei, delimiter="\t")
    csv_hate = list(csv_reader_object)[1:]
    for row in csv_hate:
        x = row[1]
        x = x.replace("#", "")
        x = re.sub(substring,"", x)
        x+="."
        x = x.lower()
        x = remove_emoji(x)
        hatespeech4 += " " + x
```

```python
hatespeech5=""

with open("hatespeech/germeval2018.training.txt") as f:
    content = f.readlines()
    for row in content:
        if "OFFENSE" in row:
            x = row
            x = re.sub(substring,"", x)
            x+="."
            x = x.lower()
            x = remove_emoji(x)
            hatespeech5 += " " + x
```

```python
hatespeech5[:100]
```

```python
hatespeech6=""

with open("hatespeech/germeval2018.test.txt") as f:
    content = f.readlines()
    for row in content:
        if "OFFENSE" in row:
            x = row
            x = re.sub(substring,"", x)
            x+="."
            x = x.lower()
            x = remove_emoji(x)
            hatespeech6 += " " + x
```

```python
hatespeech_corpora= hatespeech + hatespeech2+ hatespeech3 + hatespeech4 + hatespeech5 + hatespeech6
```

```python
sents = nltk.sent_tokenize(hatespeech_corpora, "german")
```

```python
sents
```

```python
with open('hatespeech_as_text', 'w') as f:
    f.write(hatespeech_corpora)
```

```python
def clean(text):
    text = nltk.word_tokenize(text)
    text = [word for word in text if word.isalpha()]
    text = [x for x in text if len(x) >1]
    return text
```

```python
sents_tokenized =[clean(x) for x in sents ]
```

```python
sents_tokenized
```

```python
def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f)
```

```python
save_obj(sents_tokenized, "hatespeech_sent_tokenized")
```
